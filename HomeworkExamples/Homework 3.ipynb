# Import libraries and data

```python
import pandas as pd

dfTitles = pd.read_csv('http://128.198.162.191/dataFiles/IMDB/titleBasics.tsv', sep='\t', low_memory=False)
dfRatings = pd.read_csv('http://128.198.162.191/dataFiles/IMDB/ratings.tsv', sep='\t', low_memory=False)
dfNames = pd.read_csv('http://128.198.162.191/dataFiles/IMDB/nameBasics.tsv', sep='\t', low_memory=False)
dfCrew = pd.read_csv('http://128.198.162.191/dataFiles/IMDB/crew.tsv', sep='\t', low_memory=False)
```

# Question 1: Director with the highest rating and most votes

```python
dfRatings.head(1)
```

```python
dfCrew.head(1)
```

## Merge Ratings and Crew
Initial merging of ratings and crew to get to our answer.

```python
ratingsAndCrew = pd.merge(dfRatings, dfCrew, on='tconst')

ratingsAndCrew
```

## Renaming the Directors column

```python
dfCrewRename = ratingsAndCrew.rename(columns={'directors': 'nconst'})

dfCrewRename
```

## Merging Director name information


```python
ratingsCrewDir = pd.merge(dfCrewRename, dfNames, on='nconst')

ratingsCrewDir
```

## Sorting the information in the dataframe


```python
ratingsCrewDir.sort_values(by=['averageRating', 'numVotes'], ascending=False)
```

## Output the appropriate results

```python
semi1 = (ratingsCrewDir.sort_values(by=['averageRating', 'numVotes'], ascending=False))
         
semi2 = semi1[['primaryName', 'averageRating', 'numVotes']]
         
final = (semi2.head(1))
         
final
```

## Write the information to a file

```python
final.to_csv('../files/question1.csv', index=False)
```

# Question 2: Highest rated film for the year 2000 by votes then rating

```python
dfTitles.head(1)
```

```python
dfRatings.head(1)
```

## Merge dataframes

```python
highRated = pd.merge(dfTitles, dfRatings, on='tconst')

highRated
```

## Query to remove everything except movies in the year 2000.

```python
movieFilter = (highRated.query('titleType == "movie" & startYear == "2000"'))

movieFilter
```

## Sort the values

```python
movieFilter.sort_values(by=['numVotes', 'averageRating'], ascending=False)
```

## Format for output

```python
semi1 = (movieFilter.sort_values(by=['numVotes', 'averageRating'], ascending=False))

semi2 = semi1[['startYear', 'primaryTitle', 'numVotes', 'averageRating']]

final=(semi2.head(5))

final
```

## Write to a file

```python
final.to_csv('../files/question2.csv', index=False)
```
