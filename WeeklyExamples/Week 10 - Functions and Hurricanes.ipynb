```python
# Import our libararies and data
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://www.nhc.noaa.gov/data/hurdat/hurdat2-1851-2024-040425.txt')

df.head()
```

```python
df.info()
```

# Add column names to our data

```python
# Create a variable for the file path to use in my renaming
filePath = 'https://www.nhc.noaa.gov/data/hurdat/hurdat2-1851-2024-040425.txt'

# Define custom column names in our data frame.
column_names = [
    'ID-Date', 'Time', 'RecordID', 'Status', 'Latitude', 'Longitude', 
    'WindSpeed', 'Pressure', 'ExtraField1', 'ExtraField2', 'ExtraField3', 
    'ExtraField4', 'ExtraField5', 'ExtraField6', 'ExtraField7', 'ExtraField8', 
    'ExtraField9', 'ExtraField10', 'ExtraField11', 'ExtraField12' 
]

df  = pd.read_csv(filePath, header=0, names=column_names, usecols=range(len(column_names)), on_bad_lines='skip')

df.head()
```

# Extract "ID", "YEAR", and "MONTH" from the ID-Date.

```python
# Create the columns to hold the data after extraction
df['YEAR'] = ''
df['MONTH'] = ''

# Only extract the Year and Month where "ID-Date" does NOT contain 'AL'.
df.loc[~df['ID-Date'].str.contains('AL'), 'YEAR'] = df['ID-Date'].str[:4]
df.loc[~df['ID-Date'].str.contains('AL'), 'MONTH'] = df['ID-Date'].str[4:6]

# Create a new "UNIQUE_ID" column that copies the AL identifier where applicable.
df['UNIQUE_ID'] = df['ID-Date'].apply(lambda x: x if 'AL' in x else None)

# Forward fill the Unique_ID
df['UNIQUE_ID'] = df['UNIQUE_ID'].ffill()

df.head()
```

# What were the total number of hurricanes by year?

```python
hurricanesByYear = df.groupby('YEAR')['UNIQUE_ID'].count().reset_index(name='COUNT')

hurricanesByYear
```

## Visualize the Hurrcianes per year

```python
# Convert the YEAR column is numeric, coerce errors to NaN
df['YEAR'] = pd.to_numeric(df['YEAR'], errors='coerce')

# Drop any rows where YEAR is Nan.
df_clean = df.dropna(subset=['YEAR'])

# Convert YEAR to an integer
df_clean.loc[:, 'YEAR'] = df_clean['YEAR'].astype(int)


# Group the Hurricanes
hurricanesByYear = df_clean.groupby('YEAR')['UNIQUE_ID'].nunique()


# Plot the results
plt.figure(figsize=(10,5))
plt.plot(hurricanesByYear.index, hurricanesByYear.values, marker='o')

plt.xlabel('Year')
plt.ylabel('Number of Unique Hurricanes')
plt.title('Number of unique Hurricanes per Year')

# Add some ticks
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(5))
plt.xticks(rotation = 45)

plt.grid(True)
plt.show()
```

```python
# Add an average line

# Plot the results
plt.figure(figsize=(10,5))
plt.plot(hurricanesByYear.index, hurricanesByYear.values, marker='o')

# Add an average line to the graph
cumulative_avg_hurricanes = hurricanesByYear.expanding().mean()

plt.plot(cumulative_avg_hurricanes.index, cumulative_avg_hurricanes.values, color='red', linestyle='dashed', label='Cumulative average')

plt.xlabel('Year')
plt.ylabel('Number of Unique Hurricanes')
plt.title('Number of unique Hurricanes per Year')

# Add some ticks
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(5))
plt.xticks(rotation = 45)

plt.grid(True)
plt.show()
```

# What was the total number of Hurricanes by Category and Year?

## Functions
Python functions are like sub-routines, or mini programs that perform a task based on the input they are given.

```python
df.info()
```

```python
# Create a function to return the Hurricane category
def categorize_wind_speed(speed):
    if 74 <= speed <= 95:
        return "Category 1"
    elif 96 <= speed <= 110:
        return "Category 2"
    elif 111 <= speed <= 129:
        return "Category 3"
    elif 130 <= speed <= 156:
        return "Category 4"
    elif speed >= 157:
        return "Category 5"
    
df['CATEGORY'] = df['WindSpeed'].apply(categorize_wind_speed)

# We could filter the data for certain conditions
# df = df[df['YEAR'] > 2000]
# df = df[df['CATEGORY'] == 'Category 5']

# Group our data by year and category
hurricanes_by_year_category = df.groupby(['YEAR', 'CATEGORY'])['UNIQUE_ID'].count().unstack(fill_value=0)

hurricanes_by_year_category
```

## Plot the results

```python
plt.figure(figsize=(15,6))

for category in hurricanes_by_year_category.columns:
    plt.plot(hurricanes_by_year_category.index, hurricanes_by_year_category[category], label=category)

plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.title('Number of Hurricanes by year and category')
plt.grid()
plt.legend()
plt.show()
```

```python
hurricanes_by_year_category.plot(kind="barh", stacked=True, figsize=(10, 40))

plt.ylabel('Year')
plt.xlabel('Number of Hurricanes')
plt.title('Number of Hurricanes by Year and Category')
plt.grid(axis='x')
plt.legend(title="Category")

plt.show()
```
